#**Qual Ã© a sÃ©rie mais bem avaliada?**



```python
# Encontrando o Ã­ndice da sÃ©rie com a maior nota
indice_serie_melhor_avaliada = df['Rating'].idxmax()

# Encontrando o nome da sÃ©rie com a maior nota
serie_melhor_avaliada = df.loc[indice_serie_melhor_avaliada, 'Name']

# Encontrando a nota dessa sÃ©rie
nota_melhor_avaliada = df.loc[indice_serie_melhor_avaliada, 'Rating']

print(f"A sÃ©rie mais bem avaliada Ã©: {serie_melhor_avaliada} com uma nota de {nota_melhor_avaliada}")
```

```python
# Ordenando o DataFrame pelas notas em ordem decrescente e selecionando as 5 primeiras sÃ©ries
df_top5 = df.sort_values(by='Rating', ascending=False).head(5)

# Criando o grÃ¡fico de barras com Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating', y='Name', data=df_top5, palette='viridis')

# Adicionando tÃ­tulos e rÃ³tulos
plt.title('Top 5 SÃ©ries Mais Bem Avaliadas')
plt.xlabel('Rating')
plt.ylabel('Name')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrando o grÃ¡fico
plt.show()
```

#**Qual Ã© a sÃ©rie mais mal avaliada?**

```python
# Encontrando o Ã­ndice da sÃ©rie com a maior nota
indice_serie_melhor_avaliada = df['Rating'].idxmin()

# Encontrando o nome da sÃ©rie com a maior nota
serie_melhor_avaliada = df.loc[indice_serie_melhor_avaliada, 'Name']

# Encontrando a nota dessa sÃ©rie
nota_melhor_avaliada = df.loc[indice_serie_melhor_avaliada, 'Rating']

print(f"A sÃ©rie mais mal avaliada Ã©: {serie_melhor_avaliada} com uma nota de {nota_melhor_avaliada}")
```

```python
# Ordenando o DataFrame pelas notas em ordem decrescente e selecionando as 5 primeiras sÃ©ries
df_bottom5 = df.sort_values(by='Rating', ascending=True).head(5)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Rating', y='Name', data=df_bottom5, palette='viridis')

# Adicionando os valores das notas no topo de cada barra
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', label_type='edge')

# Adicionando tÃ­tulos e rÃ³tulos
plt.title('Top 5 SÃ©ries Mais Mal Avaliadas')
plt.xlabel('Nota')
plt.ylabel('SÃ©rie')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrando o grÃ¡fico
plt.show()
```

#**Qual a relaÃ§Ã£o entre os dias da semana e as sÃ©ries mais bem avaliadas?**



```python
# Agrupando por dia da semana e calculando a mÃ©dia das notas
df_grouped = df.groupby('Schedule (days)')['Rating'].mean().reset_index()

# Ordenando por dia da semana para garantir a ordem correta
dias_semana = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_grouped['Schedule (days)'] = pd.Categorical(df_grouped['Schedule (days)'], categories=dias_semana, ordered=True)
df_grouped = df_grouped.sort_values('Schedule (days)')

# Criando o grÃ¡fico de barras com Seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Schedule (days)', y='Rating', data=df_grouped, palette='viridis')

# Adicionando os valores das notas no topo de cada barra
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                textcoords='offset points')

# Adicionando tÃ­tulos e rÃ³tulos
plt.title('MÃ©dia das Notas das SÃ©ries por Dia da Semana de ExibiÃ§Ã£o')
plt.xlabel('Dia da Semana')
plt.ylabel('Nota MÃ©dia')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrando o grÃ¡fico
plt.show()
```
